
Write a program to do following: We have given a collection of 8 points. P1=[0.1,0.6] P2=[0.15,0.71] P3=[0.08,0.9] P4=[0.16, 0.85] P5=[0.2,0.3] P6=[0.25,0.5] P7=[0.24,0.1] P8=[0.3,0.2]. Perform the k-mean clustering with initial centroids as m1=P1=Cluster#1=C1 and m2=P8=cluster#2=C2. Answer the following: a) Which cluster does P6 belong to? b) What is the population of a cluster around m2? c) What is the updated value of m1 and m2?

```python
import numpy as np
points = {
    'P1': [0.1, 0.6],
    'P2': [0.15, 0.71],
    'P3': [0.08, 0.9],
    'P4': [0.16, 0.85],
    'P5': [0.2, 0.3],
    'P6': [0.25, 0.5],
    'P7': [0.24, 0.1],
    'P8': [0.3, 0.2]
}
```

```python
m1 = np.array(points['P1'])
m2 = np.array(points['P8'])
```

```python
C1 = []
C2 = []
```

```python
def euclidean_distance(p1, p2):
    return np.sqrt(np.sum((p1 - p2) ** 2))
```

```python
for point_name, point_coords in points.items():
    point = np.array(point_coords)
    distance_to_m1 = euclidean_distance(point, m1)
    distance_to_m2 = euclidean_distance(point, m2)

    if distance_to_m1 < distance_to_m2:
        C1.append(point)
    else:
        C2.append(point)
```

```python
new_m1 = np.mean(C1, axis=0)
new_m2 = np.mean(C2, axis=0)
```

```python
if np.array_equal(new_m1, m1):
    cluster_p6 = "C1"
else:
    cluster_p6 = "C2"
```

```python
print("a) P6 belongs to Cluster:", cluster_p6)
```

```python
population_m2 = len(C2)
```

```python
print("b) Population of cluster around m2:", population_m2)
```

```python
print("c) Updated value of m1:", new_m1)
print("   Updated value of m2:", new_m2)
```
